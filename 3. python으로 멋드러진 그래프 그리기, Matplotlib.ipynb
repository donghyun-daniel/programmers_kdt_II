## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- `%matplotlib inline` 이제는 안쳐도 default :)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## II. Matplotlib Case Study with Arguments

```python
plt.figure(figsize = (8, 6)) # plotting을 할 도면을 선언 (figsize의 인자는 (가로, 세로))
plt.plot([1, 2, 3, 4, 5])
plt.show() # plt 확인 명령
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해 1차 함수 y=x를 그려보면:
plt.plot([i for i in range(5)])
plt.show()
```

```python
#np.array를 이용해 함수 그래프 만들기

x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) -> c를 float로 넣어줘도 된다!
x = np.arange(-10, 10, 0.01)

plt.figure(figsize = (10, 6)) # plotting을 할 도면을 선언 (figsize의 인자는 (가로, 세로))

# x, y축에 설명 추가하기
plt.xlabel("x value")
plt.ylabel("f(x) value")

# x, y축의 눈금 설정하기
plt.xticks([i for i in range(-5, 5, 1)])
plt.yticks([i for i in range(0, 26, 2)])

# x, y축의 범위를 설정하기
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]

# 그래프에 title 달기
plt.title("y = x^2 graph")

plt.plot(x, x ** 2, label = "trend") # label은 범례 옵션

plt.legend() # legend()는 범례 생성 .plot()보다 뒤에 나와야한다

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프 (Plot)

```python
x = np.arange(20)
y = np.random.randint(0, 20, 20) # (a, b, c)에서 a, b는 범위, c는 난수 생성 횟수

plt.plot(x, y)

#y축을 20까지 보이게 하고 싶다면
plt.axis([-1, 20, 0, 18])
plt.show()
```

### 산점도 (Scatter Plot)

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot((x, y))
plt.title("Box Plot of y")
plt.show()
```

맨위 맨아래의 선이 min, max, 가운데 사각형의 아래변이 Q1, 주황선이 Q2, 윗변이 Q3에 해당

### 막대 그래프 (Bar Plot)
- 범주형 데이터의 "값"과 그 값을 크기를 직사각형으로 나타낸 그림

```python
plt.bar(y, x)
plt.xticks(np.arange(0, 22, 1))
plt.show()
```

### 히스토그램 (Histogram)

- 막대그래프와의 공통점 : 도수분포를 직사각형의 막대 형태로 나타낸다
- 차이점 : 히스토그램은 변량을 모두 표시하기보다 변량을 그룹화한 "계급"으로 나타낸 것이 특징이다.
- ex) 0, 1, 2가 아니라, 0 ~ 2 까지의 "범주형 데이터"로 구성 후 그림을 그린다

```python
plt.hist(y, x)
plt.xticks(np.arange(0, 22, 2))
plt.show()
```

### 원형 그래프 (Pie Chart)
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해 **비율**을 확인하는데 좋다!

```python
z = [100, 300, 200, 400]

plt.pie(z, labels = ["one", "two", "three", "four"])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Matplotlib 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도그림 (Kernel Density Plot)
- 카운트그림 (Count Plot)
- 캣그림 (Cat Plot)
- 스트립그림 (Strip Plot)
- 히트맵 (Heatmap)

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화 해서 그린 그림
- `sns.kdeplot()`

```python
# in Histogram(0, 22, 2)

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins = x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade = True) # shade는 그래프 아래 부분에 대해 음영을 boolean으로 설정
plt.show()
```

### 카운트그림 (Count Plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수 확인하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name" : ["Andy", "Bob", "Cat"], "vote" : [True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby("vote").count()
vote_count
```

```python
plt.bar(x = [False, True], height = vote_count["name"]) # x = [False, True] 는 0은 False, 1은 True라는 의미
plt.show()
```

```python
# sns의 countplot

sns.countplot(x = vote_df["vote"])
plt.show()
```

### 캣그림 (Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`
- 복잡한 데이터에서 더 좋다 :)

```python
covid = pd.read_csv("./country_wise_latest.csv")
covid.head(5)
```

```python
# WHO Region 별 확진자

# hue param을 추가해줄수도 있다, kind 인자를 안넣으면 "strip"이 default다
s = sns.catplot(x = "WHO Region", y = "Confirmed", data = covid, kind = "violin") 
s.fig.set_size_inches(10, 6) #sns 스타일의 사이즈 설정
plt.show()
```

### 스트립그림 (Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x = "WHO Region" , y = "Recovered", data = covid)
plt.show()
```

```python
# cf) swarmplt
# 비슷한 값이 뭉쳐서 있으면 산점도의 형태는 조금 보기 안좋을 수 있다

sns.swarmplot(x = "WHO Region" , y = "Recovered", data = covid)
plt.show()
```

### 히트맵 (Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`
- 가장 많이 쓰는 경우는 correlation 행렬이다 :)

```python
# 히트맵 예제
covid.corr()
```

```python
sns.heatmap(covid.corr())
plt.show()
```
